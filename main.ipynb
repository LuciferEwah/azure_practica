```python
try:
    # Establecer la conexión
    conexion = pyodbc.connect(connection_string)
    cursor = conexion.cursor()

    consulta_dataframe = "SELECT * FROM Venta"
    cursor.execute(consulta_dataframe)

    # Obtener los nombres de las columnas desde el cursor
    nombres_columnas = [columna[0] for columna in cursor.description]

    # Inicializar un diccionario para almacenar los datos
    diccionario_datos = {nombre_columna: [] for nombre_columna in nombres_columnas}

    # Iterar sobre cada fila y almacenar los valores en el diccionario
    fila = cursor.fetchone()
    while fila:
        for nombre_columna, valor in zip(nombres_columnas, fila):
            diccionario_datos[nombre_columna].append(valor)
        fila = cursor.fetchone()

    # Crear el DataFrame desde el diccionario
    dataframe = pd.DataFrame(diccionario_datos)

    cursor.close()
    conexion.close()
    
except pyodbc.Error as error:
    print("Ha ocurrido un error al intentar conectarse a la base de datos:", error)
```

```python

fechas = pd.date_range(start='2010-01-01', end='2030-12-31', freq='D')

# Crear DataFrame para la dimensión de tiempo
dim_tiempo = pd.DataFrame({
    'FECHA': fechas,
    'AÑO': fechas.year,
    'MES': fechas.month,
    'DIA': fechas.day,
    'DIA_SEMANA': fechas.dayofweek,
    'NOMBRE_MES': fechas.strftime('%B'),  # Nombre del mes
    'NOMBRE_DIA_SEMANA': fechas.strftime('%A')  # Nombre del día de la semana
})

# Agregar un ID único para la dimensión de tiempo
dim_tiempo['ID_TIEMPO'] = range(1, len(dim_tiempo) + 1)

dim_tiempo['NOMBRE_MES'] = dim_tiempo['NOMBRE_MES'].replace({
    'January': 'Enero', 'February': 'Febrero', 'March': 'Marzo', 'April': 'Abril',
    'May': 'Mayo', 'June': 'Junio', 'July': 'Julio', 'August': 'Agosto',
    'September': 'Septiembre', 'October': 'Octubre', 'November': 'Noviembre', 'December': 'Diciembre'
})
dim_tiempo['NOMBRE_DIA_SEMANA'] = dim_tiempo['NOMBRE_DIA_SEMANA'].replace({
    'Monday': 'Lunes', 'Tuesday': 'Martes', 'Wednesday': 'Miércoles', 'Thursday': 'Jueves',
    'Friday': 'Viernes', 'Saturday': 'Sábado', 'Sunday': 'Domingo'
})
dim_tiempo['ID_TIEMPO'] = range(1, len(dim_tiempo) + 1)
```

```python

def quitar_tildes_en_dataframe(df):
    for columna in df.columns:
        # Verificar si el valor es un string antes de aplicar unidecode
        df[columna] = df[columna].apply(lambda x: unidecode(x).replace('Ñ', '<NH>').replace('ñ', '<nh>') if isinstance(x, str) else x)


# Aplicar la función a cada DataFrame
quitar_tildes_en_dataframe(fact_venta)
quitar_tildes_en_dataframe(dim_tipo_compra)
quitar_tildes_en_dataframe(dim_producto)
quitar_tildes_en_dataframe(dim_tiempo)

# Guardar los DataFrames en archivos CSV
dim_producto.to_csv('dim_producto.csv', index=False, encoding='utf-8')
dim_tipo_compra.to_csv('dim_tipo_compra.csv', index=False, encoding='utf-8')
fact_venta.to_csv('fact_venta.csv', index=False, encoding='utf-8')
dim_tiempo.to_csv('dim_tiempo.csv', index=False, encoding='utf-8')

```
